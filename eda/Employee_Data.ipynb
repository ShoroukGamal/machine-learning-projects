```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("Employee_Data.xls")
```

```python
df.head()
```

```python
df.drop(["Employee Id", "First Name", "Last Name"], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df["Department"].value_counts()
```

```python
sns.displot(df.Salary, color="green")          #distribution plot  (range)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Experience", y="Salary")           #help me to understand the outlier
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Age", y="Salary")
plt.show()
```

```python
sns.countplot(x="Department", hue="Department", data=df)     #to count the classification,  hue like legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x="Age", hue="Age", data=df)   #count  relative to hue......the result y is the count of x
plt.show()
```

```python
sns.barplot(x="Department", y="Salary",hue= "Department", data=df)
plt.show()
```

```python
sns.boxplot(df["Experience"])
plt.show()
```

```python
#strip plot
#group the strips by categorical variable

sns.stripplot(x="Department", y="Salary", data=df)
plt.show()
```

```python
data_1 = df[["Age","Experience","Salary"]]
```

```python
corr_matrix = data_1.corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
sns.lmplot(x="Age", y="Salary", data=df)       #linear model
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.regplot(x="Age", y="Salary", data=df, marker= 'x')         #regression plot
plt.show()
```
