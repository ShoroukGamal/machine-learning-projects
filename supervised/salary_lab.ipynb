```python
df_new = df_new.drop(["relationship"], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in ["workclass","marital-status", "occupation", "race", "sex", "native-country", "salary"]:
    df_new[col] = le.fit_transform(df_new[col])
```

```python
df_new["salary"].unique()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
for col in ["age", "fnlwgt", "hours-per-week"]:
    df_new[col] = scaler.fit_transform(np.array(df_new[col]).reshape(-1,1))
```

```python
df_new["age"].min(), df_new["age"].max()
```

```python
df_new["fnlwgt"].min(), df_new["fnlwgt"].max()
```

```python
df_new["hours-per-week"].min(), df_new["hours-per-week"].max()
```

```python
x = df_new.drop(["salary"], axis=1)
y = df_new["salary"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(x_train, y_train)

print("Train Score: ", model.score(x_train, y_train))
print("Test Score: ", model.score(x_test, y_test))
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred)
print(classification_report(y_test, y_pred))
```

```python
sns.heatmap(cm, annot=True,fmt= ' .10g')
plt.show()
```
